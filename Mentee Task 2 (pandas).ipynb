## average tempo for each of the top 10 most occurring artists.

```python
top_10_artists = list(df['artist'].value_counts().head(10).index)
top_10_artists
```

```python
average_tempo = []
for i in top_10_artists:
    tempo_avg = df[df['artist'] == i]['tempo'].mean()
    average_tempo.append({'Artist': i, 'Average Tempo': tempo_avg})
    
df2 = pd.DataFrame(average_tempo)
df2
```
